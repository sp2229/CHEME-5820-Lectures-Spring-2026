# L4c: Covariance Redux - Linear Kernels and Beyond
In L2a we derived the empirical covariance matrix by centering data and taking inner products. In kernel methods, that same algebra appears again as a Gram matrix built from a kernel. Let's make the relationship precise and then generalize it.

We will use the following notation:
* $\mathbf{x}_i \in \mathbb{R}^m$ denotes sample $i$ (a row vector of $m$ features)
* $\mathbf{X}\in\mathbb{R}^{n\times m}$ stacks the samples by rows
* $\mathbf{1}\in\mathbb{R}^n$ is the vector of ones, and $\mathbf{H}=\mathbf{I}-\frac{1}{n}\mathbf{1}\mathbf{1}^\top$ is the centering matrix
* $\mathbf{X}_c=\mathbf{H}\mathbf{X}$ is the centered data matrix, and $\bar{\mathbf{x}}=\frac{1}{n}\sum_{i=1}^n \mathbf{x}_i$ is the sample mean

This is consistent with L2a: the empirical covariance is computed across samples and lives in feature space.


## Empirical covariance as a linear-kernel Gram matrix
The empirical covariance matrix is
$$
\hat{\mathbf{\Sigma}} = \frac{1}{n-1}\mathbf{X}_c^\top \mathbf{X}_c.
$$
Entry-wise, for features $i$ and $j$,
$$
\hat{\Sigma}_{ij} = \frac{1}{n-1}\sum_{k=1}^n (x_k^{(i)}-\bar{x}_i)(x_k^{(j)}-\bar{x}_j).
$$

Define the linear kernel
$$
k(\mathbf{x}_i,\mathbf{x}_j)=\mathbf{x}_i^\top \mathbf{x}_j.
$$

If we apply this kernel to the **centered feature vectors** (columns of $\mathbf{X}_c$), we obtain the feature-space Gram matrix:
$$
\mathbf{G}_{\text{lin}}=\mathbf{X}_c^\top \mathbf{X}_c.
$$

Therefore,
$$
\boxed{\hat{\mathbf{\Sigma}}=\frac{1}{n-1}\mathbf{G}_{\text{lin}}}
$$

> __Why two Gram matrices?__
>
> The same data matrix produces two linear-kernel Gram matrices:
> * $\mathbf{G}_{\text{lin}}=\mathbf{X}_c^\top\mathbf{X}_c\in\mathbb{R}^{m\times m}$ compares **features** (this is the covariance up to scaling).
> * $\mathbf{K}_{\text{lin}}=\mathbf{X}_c\mathbf{X}_c^\top\in\mathbb{R}^{n\times n}$ compares **samples** (this is the standard kernel matrix used in kernel methods).
>
> These matrices share the same nonzero eigenvalues (a direct SVD consequence), but their eigenvectors live in different spaces.


## Centering a kernel matrix
Kernel methods usually start from the sample-space kernel matrix:
$$
\mathbf{K}_{ij}=k(\mathbf{x}_i,\mathbf{x}_j).
$$
To match covariance, we must center in feature space. The centered kernel matrix is
$$
\mathbf{K}_c=\mathbf{H}\mathbf{K}\mathbf{H},\qquad \mathbf{H}=\mathbf{I}-\frac{1}{n}\mathbf{1}\mathbf{1}^\top.
$$

For the linear kernel,
$$
\mathbf{K}_c=\mathbf{H}(\mathbf{X}\mathbf{X}^\top)\mathbf{H}=(\mathbf{H}\mathbf{X})(\mathbf{H}\mathbf{X})^\top=\mathbf{X}_c\mathbf{X}_c^\top,
$$
so centering the kernel matrix is exactly the same as centering the data matrix first.

**Note:** Without centering, the linear kernel captures second moments about the origin, not covariance about the sample mean.


## Beyond linear: covariance in a feature space
Every positive semidefinite kernel corresponds to an inner product in some feature space:
$$
k(\mathbf{x}_i,\mathbf{x}_j)=\langle \phi(\mathbf{x}_i),\phi(\mathbf{x}_j)\rangle.
$$

Define the feature-space mean
$$
\mu_\phi=\frac{1}{n}\sum_{i=1}^n \phi(\mathbf{x}_i)
$$
and the feature-space covariance operator
$$
\mathbf{C}=\frac{1}{n-1}\sum_{i=1}^n (\phi(\mathbf{x}_i)-\mu_\phi)(\phi(\mathbf{x}_i)-\mu_\phi)^\top.
$$

The centered kernel matrix encodes exactly the inner products of these centered features:
$$
(\mathbf{K}_c)_{ij}=\langle \phi(\mathbf{x}_i)-\mu_\phi,\ \phi(\mathbf{x}_j)-\mu_\phi\rangle.
$$

This is the kernel version of covariance. Kernel PCA uses the eigendecomposition of $\mathbf{K}_c$ to recover principal directions in feature space. With the linear kernel, this reduces to ordinary covariance and PCA.


## Summary
We can view the empirical covariance matrix as a special case of a kernel Gram matrix, and kernels as covariance in a transformed feature space.

> __Key Takeaways:__
>
> * $\hat{\mathbf{\Sigma}}=\frac{1}{n-1}\mathbf{X}_c^\top\mathbf{X}_c$ is a scaled linear-kernel Gram matrix on centered data.
> * The sample-space kernel matrix is $\mathbf{K}_{ij}=k(\mathbf{x}_i,\mathbf{x}_j)$, and the centered kernel is $\mathbf{K}_c=\mathbf{H}\mathbf{K}\mathbf{H}$.
> * Any valid kernel defines an implicit feature map $\phi$; $\mathbf{K}_c$ captures covariance in that feature space.

The kernel trick keeps the same algebra while changing the similarity function. The linear kernel recovers standard covariance exactly.
